## 기초 통계 계산하기

### lifeExp 열을 연도별로 그룹화하여 평균 계산하기

```python
df.groupby('year')['lifeExp'].mean() # 앗 어렵다.. 의미를 이해하자
```

```python
# 데이터프레임 형태로 출력하려면 어떻게 해야할까?

df.groupby('year')[['lifeExp']].mean()
```

```python
df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean() 
# 연도와 대륙 기준으로 그룹핑을 하고, 평균수명과 gdp의 평균을 구하는 것
```

### (2) 데이터프레임 만들기

```python
# 딕셔너리를 DataFrame클래스에 전달하여 딕셔너리 생성

scientists = pd.DataFrame({
    'Name': ['Rosaline Franklin', 'William Gosset'],
    'Occupation': ['Chemist', 'Statistician'],
    'Born': ['1920-04-16', '1876-06-13'],
    'Died': ['1958-04-16', '1937-10-16'],
    'Age': [37, 61]
})
print(scientists)

# 0과 1 인덱스가 자동으로 생성됨을 알 수 있다.
```

## 데이터프레임 다루기

### scientists.csv 파일 임포트

```python
# (예시) scientists = pd.read_csv('/gdrive/My Drive/220205/scientists.csv')

# 나의 절대 로 작성
scientists = pd.read_csv('/gdrive/My Drive/Pandas_practice/scientists.csv')

scientists.head()
```

### 통계내기

```python
# 나이의 최소값
print(scientists['Age'].min())
# 나이의 최대값
print(scientists['Age'].max())
# 나이의 평균
print(scientists['Age'].mean())
```

### 조건을 만족하는 행만 추출하기

```python
# 평균보다 나이가 어리다는 조건을 만족하는 행만 추출

scientists[scientists['Age'] < scientists['Age'].mean()] 
```

## 데이터프레임의 데이터 처리하기

### 날짜를 datetime 자료형으로 바꾸기

Born과 Died 열은 Object(문자열)이다. 날짜이므로 datetime이라는 자료형으로 바꿔보자.

```python
print(scientists['Born'].dtype)
print(scientists['Died'].dtype)
```

```python
# Born 열을 datetime 자료형으로 바꿔서 born_datetime 변수에 저장하기
born_datetime = pd.to_datetime(scientists['Born'], format='%Y-%m-%d')
print(born_datetime)
```

```python
# Died 열을 datetime 자료형으로 바꿔서 died_datetime 변수에 저장하기
died_datetime = pd.to_datetime(scientists['Died'], format='%Y-%m-%d')
print(died_datetime)
```

### 새로운 열 추가하기

datetime으로 바꾼 것을 scientists 데이터프레임에 새로운 열로 추가한다

```python
# born_dt, died_dt 추가
scientists['born_dt'] = born_datetime
scientists['died_dt'] = died_datetime
scientists.head()
```

### 시간 계산하기

died_dt열에서 born_dt를 빼면 살아있던 시간을 계산할 수 있다.

```python
# 과학자들이 살아있던 날 계산하여 열에 추가
scientists['age_days_dt'] = scientists['died_dt'] - scientists['born_dt']
scientists.head()
```

### 열 삭제하기

열을 통째로 삭제하려면 drop() 함수를 사용한다.

```python
# age_days_dt 열 삭제
scientists_dropped = scientists.drop(['age_days_dt'], axis=1) # axis = 1 : 열 방향으로 동작한다 / axis = 0 : 행 방향으로 동작한다
scientists_dropped
```

```python
scientists.head()
```

```python
scientists.drop(['age_days_dt'], axis=1, inplace = True)
# axis = 1 : 열 방향으로 동작한다 / axis = 0 : 행 방향으로 동작한다
# inplace = True : 데이터 프레임 원본 자체에서 삭제할 수 있도록 하는 옵션
```

```python
scientists.head()
```
